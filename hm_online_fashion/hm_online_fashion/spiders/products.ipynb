```python
import pandas as pd
import json
```

```python
with open('./products.json', encoding='utf-8') as f:
    data = json.load(f)
    for element in data:
        if type(element['other_img_colors_products']) == dict:
            element['other_img_colors_products'] = [element['other_img_colors_products']]
        if element['other_img_colors_products'][0]: 
            keys = element['other_img_colors_products'][0].keys()
            for key in keys:
                element['other_img_colors_products'][0][key] = f'https:{element["other_img_colors_products"][0][key]}'
        if element['image_link']:
            element['image_link'] = f'https:{element["image_link"]}'

    df = pd.DataFrame(data)

df

```
