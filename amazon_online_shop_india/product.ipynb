```python
import pandas as pd
import json
```

```python
df = pd.read_csv("./products.csv")
df
```

```python
with open('./products.json') as f:
    data = json.loads(f.read())
    for item in data:
        if item['product_details']:
            item['product_details'] = [item['product_details']]
    

new_df = pd.json_normalize(data)
new_df
```

```python
new_df.to_csv('products.csv')
```
